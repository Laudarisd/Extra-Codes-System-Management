```python
 %matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
import os
```

```python
df1 = pd.read_csv('./original_data/1.csv')
df2 = pd.read_csv('./original_data/2.csv')


dir = "./figure"

if not os.path.exists(dir):
    os.mkdir(dir)
#df2 = pd.read_csv('./data/B.csv')
import prettyplotlib as ppl
fig, ax = plt.subplots(1)
for i in range(1):
    x_1 = df1["y"][1000:]
    y_1 = df1["z"][1000:]
    #x = np.random.normal(loc=i, size=1000)
    #y = np.random.normal(loc=i, size=1000)
    ax.scatter(x_1, y_1, label='group A', facecolor='red', edgecolor='black', linewidth=1)
    #ax.xticks([1, 1000])

for j in range(1):
    x_2 = df2["y"][1000:]
    y_2 = df2["z"][1000:]
    #x = np.random.normal(loc=i, size=1000)
    #y = np.random.normal(loc=i, size=1000)
    ax.scatter(x_2, y_2, label='groupB', facecolor='silver', edgecolor='black', linewidth=1)

# Get back the top and right axes lines ("spines")
spines_to_remove = ['top', 'right']
for spine in spines_to_remove:
    ax.spines[spine].set_visible(True)
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')

# For all the spines, make their line thicker and return them to be black
all_spines = ['top', 'left', 'bottom', 'right']
for spine in all_spines:
    ax.spines[spine].set_linewidth(1.0)
    ax.spines[spine].set_color('black')

# Change the axis title also back to black
ax.title.set_color('black')
ax.legend()
ax.grid(False)
#ax.set_xticks([1, 10])
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_xlabel('Position on y axis')
ax.set_ylabel("Position on z axis")
#ax.axis('off')

ax.set_title('Trajectories of A & B')
fig.savefig(os.path.join(dir,'both_yz.png'))
```
