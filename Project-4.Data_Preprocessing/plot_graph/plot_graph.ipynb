```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df_1 = pd.read_csv("CR.csv")
df1 = df_1.sort_values(by='ratio(B/A)', ascending=True)
df_2 = pd.read_csv("./radius.csv")
df2 = df_2.sort_values(by='ratio(B/A)', ascending=True)
df_3 = pd.read_csv("./YM.csv")
df3 = df_3.sort_values(by='ratio', ascending=True)
df_4 = pd.read_csv("./CF.csv")
df4 = df_4.sort_values(by='ratio(B/A)', ascending=True)
df_5 = pd.read_csv("./different_all.csv")
df5 = df_5.sort_values(by='ratio(A/B)', ascending=True)
df_6 = pd.read_csv("./density.csv")
df6 = df_6.sort_values(by='ratio(A/B)', ascending=True)
# Data
# Data
df=pd.DataFrame({'x_values': df2['range'], 'y1_values': df1["Accuracy%"], 'y2_values': df2["Accuracy%"],'y3_values': df3["Accuracy%"],'y4_values': df4["Accuracy%"],'y5_values': df5["Accuracy%"], 'y6_values': df6["Accuracy%"]})
#df=pd.DataFrame({'x_values': df2['ratio(B/A)'], 'y2_values': df2["Accuracy%"] })
#df=pd.DataFrame({'x_values': range(1,11), 'y1_values': np.random.randn(10)})
 
# multiple line plots
plt.plot( 'x_values', 'y1_values', data=df, marker='o', markerfacecolor='blue', markersize=5, color='skyblue',linestyle='dashed', linewidth=2, label = "CR")
plt.plot( 'x_values', 'y2_values', data=df, marker='o', color='olive',markersize = 5, linewidth=2,linestyle='dashed', label = "Radius")
plt.plot( 'x_values', 'y3_values', data=df, marker='o', color='red',markersize = 5, linewidth=2, linestyle='dashed', label="YM")
plt.plot( 'x_values', 'y4_values', data=df, marker='o', color='green',markersize = 5, linewidth=2, linestyle='dashed', label="CF")
plt.plot( 'x_values', 'y5_values', data=df, marker='o', color='orange',markersize = 5, linewidth=2, linestyle='dashed', label="All_diff")
plt.plot( 'x_values', 'y6_values', data=df, marker='o', color='black',markersize = 5, linewidth=2, linestyle='dashed', label="Density")
# show legend
plt.xlabel('Ratio')
plt.ylabel('Accuracy %')
plt.title("Visualization of our result")
plt.legend()
plt.savefig("./final_plt.png")
# show graph
plt.show()

```
